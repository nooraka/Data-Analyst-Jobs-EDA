## Importing required packages

```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('deep')

import warnings
warnings.filterwarnings('ignore') 
```

## Loading data into Dataframe

```python
df = pd.read_csv('DataAnalyst.csv')
df_copy = df.copy(deep=True)
df.head(3)
```

```python
df.tail()
```

## Data Pre-processing

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

Removed unnecessary column.

```python
print(df.shape)
```

Total record in dataset = 2253<br> Columns in the dataset = 15

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.info()
```

```python
df.isna().sum()
```

Checking for null values.

```python
df.replace('-1', np.nan, inplace=True)
df.replace(-1, np.nan, inplace=True)
df.replace(-1.0, np.nan, inplace=True)
```

There are some negative values, I changed them to NaN. Let's check null values again:

```python
df.isna().sum()
```

We had huge amount of negative values.<br> Let's clean the data. We start with "company_name" column.

```python
df[df['company_name'].isna()]
```

```python
df.drop(index=df.index[1860], axis=0, inplace=True)
```

The missing value in "company_name" column is removed.

```python
df.duplicated().sum()
```

No duplicate value found in features.

```python
df['job_title'].value_counts().head(10)
```

Some spelling errors in "job_title".

```python
changes_in_table = {'Sr. Data Analyst' : 'Senior Data Analyst', 
                    'Sr Data Analyst' : 'Senior Data Analyst',
                    'Data Analyst Junior' : 'Junior Data Analyst'}

df['job_title'] = df['job_title'].map(changes_in_table).fillna(df['job_title']) 
```

```python
df["job_title"].value_counts().head(10)
```

```python
df.columns = df.columns.str.lower().str.replace('Sr.', 'Senior')
```

We have to create 3 new columns from "salary_estimate" for ease of calculation.

```python
df['salary_estimate'].head(3)
```

```python
df['salary_minimum']= df['salary_estimate'].str.lstrip('$').str[:3].str.replace('K','').str.strip().astype('float')
```

```python
df['salary_maximum'] = df['salary_estimate'].str[6:10].str.replace('K','').str.lstrip('$').str.strip().replace('', float('NaN')).astype('float')
```

```python
df['salary_average'] = (df['salary_maximum']+df['salary_minimum'])/2
```

```python
df[['salary_maximum', 'salary_minimum', 'salary_average']].head()
```

```python
df['job_description'].nunique()
```

"job_description" has long text, we take it for granted.

```python
pd.DataFrame(df["company_name"].head())
```

As you can see, the rating score is duplicated in the company name. So we need to remove that from the company name:

```python
df["company_name"] = df['company_name'].str[:-4]
```

```python
pd.DataFrame(df.company_name.head())
```

```python
df['size'] = df['size'].replace(' employees', '', regex=True)
```

```python
df['size'].unique()
```

We have changed the "size" syntax.

```python
df.revenue.unique()
```

```python
df['revenue'] = df['revenue'].str.replace('$', '')
```

```python
df = df[['job_title', 'job_description',  'salary_minimum', 'salary_maximum', 'salary_average', 'rating', 'company_name', 'location', 'headquarters', 'size', 'founded', 'type_of_ownership', 'industry', 'sector', 'revenue', 'competitors', 'easy_apply']]
```

```python
df.head(3)
```

```python
df.describe().T
```

Finally, we have finished pre-processing the dataset, now we can start analyzing the data.

## Univariate Analysis

**Job Titles**

```python
df['job_title'].nunique()
```

There are 1269 job titles in our dataset.

```python
pd.DataFrame(df['job_title'].str.contains("Data Analyst").value_counts())
```

job_title column contains the word "Data Analyst" 1368 times: "Senior Data Analyst", "Junior Data Analyst", "Business Data Analyst", "Data Quality Analyst", "Data Management Analyst", etc.

```python
pd.DataFrame(df['job_title'].value_counts().nlargest(10))
```

```python
job_titles = df['job_title'].value_counts().nlargest(10).reset_index()
sns.barplot(y="index", x="job_title", data=job_titles, dodge=True);
```

Top 10 job titles.

**Job Description**

```python
df['job_description'].nunique()
```

There are 2252 different job descriptions in our dataset.

The data analyst should have the following skills: Python, SQL, MS Excel, Tableau/PowerBI. We will study those job descriptions where these skills are.

```python
skills =['Python','SQL','MS Excel','Tableau','PowerBI']

def matcher(x):
    for i in skills:
        if i.lower() in x.lower():
            return i
    else:
        return np.nan

df['DA_skills'] = df['job_description'].apply(matcher)
```

```python
pd.DataFrame(df['DA_skills'].value_counts())
```

```python
pd.DataFrame(df['DA_skills'].value_counts()).plot(kind='barh');
```

As we see, SQL and Python are trending.

**Location**

```python
df.groupby('location').DA_skills.count().reset_index().sort_values(by='DA_skills', ascending=False).nlargest(10, columns="DA_skills")
```

Most data analytics jobs are located in New York, NY.

**Salary**

```python
pd.DataFrame(df.groupby('job_title')['salary_average'].sum().reset_index().nlargest(5, columns='salary_average'))
```

Top 5 average salaries.

```python
plt.figure(figsize=[8, 4])
sns.boxplot(data=df, x="DA_skills", y="salary_average", hue="DA_skills")
plt.title("Aerage Salary for Data Analyst Skills");
```

Python, SQL and Tableau are valued above other skills.

```python
df.groupby('salary_maximum').DA_skills.describe().reset_index().sort_values(by='salary_maximum', ascending=False).nlargest(10, columns='salary_maximum')
```

Top 10 data analysts with maximum salary.<br> The following table shows the minimum wages:

```python
df.groupby('salary_minimum').DA_skills.describe().reset_index().sort_values(by='salary_minimum', ascending=False).nlargest(10, columns='salary_minimum')
```

**Rating of Jobs**

Most jobs are rated between 3 and 4.
